```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/flower.webp", cv2.IMREAD_GRAYSCALE)
laplacian_var=cv2.Laplacian(img,cv2.CV_64F).var()
print("Laplacian variance:",int(laplacian_var))
if laplacian_var<5:
    print("Image is blurry")
else:
    print("Image is not blurry")
cv2_imshow(img)
```
